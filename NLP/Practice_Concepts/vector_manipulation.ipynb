```python
import numpy as np

# Dummy word vectors for explanation
word_vectors = {
    'king': np.array([0.8, 0.6]),
    'queen': np.array([0.7, 0.7]),
    'man': np.array([0.9, 0.1]),
    'woman': np.array([0.6, 0.8])
}

# Example vector
king_vector = word_vectors['king']
```

```python
import matplotlib.pyplot as plt

# Rotation matrix for 45 degrees counterclockwise
theta = np.deg2rad(45)
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta), np.cos(theta)]])

# Rotate "king" vector
king_rotated = np.dot(R, king_vector)

# Plot original and rotated vectors
def plot_vectors(vectors, colors=['k', 'r'], labels=['Original', 'Rotated']):
    plt.figure()
    ax = plt.gca()
    for i, vec in enumerate(vectors):
        ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=labels[i])
    ax.set_xlim([-1, 1.5])
    ax.set_ylim([-1, 1.5])
    plt.legend()
    plt.draw()
    plt.show()

plot_vectors([king_vector, king_rotated])
```

```python
# Assume we have transformation matrix T trained to map English embeddings to French
T = np.array([[0.9, 0.1], [0.1, 0.9]])

# Translate "king" vector to French embedding space
king_french = np.dot(T, king_vector)

# Plot original and translated vectors
plot_vectors([king_vector, king_french], colors=['k', 'g'], labels=['English', 'French'])
```
